```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Initialize the Data

```python
data = {
    'Company_ID': np.arange(1, 13),
    'Company_Name': ['Apple', 'Microsoft', 'Google', 'Amazon', 'Meta', 'IBM', 'Oracle', 'Salesforce', 'Adobe', 'Intel', 'Tesla', 'NVIDIA'],
    'Domain': ['Hardware', 'Software', 'Search Engine', 'E-commerce', 'Social Media', 'AI & Cloud', 'Database', 'CRM', 'Creative Software', 'Semiconductors', 'EV', 'GPUs'],
    'Stock_Price': np.round(np.random.uniform(100, 2000, 12), 2)
}

```

# Creating DataFrame

```python
df = pd.DataFrame(data)
# Display the first few rows of the DataFrame
display(df)
```

# Scatter plot

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Company_ID'], df['Stock_Price'], color='blue', alpha=0.6)
plt.xlabel('Company ID')
plt.ylabel('Stock Price')
plt.title('IT Company Stock Prices')
plt.grid(True)
plt.show()
```

# Bar chart

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Company_Name'], df['Stock_Price'], color='green', alpha=0.7)
plt.xlabel('Company Name')
plt.ylabel('Stock Price')
plt.title('Stock Price per IT Company')
plt.xticks(rotation=45)
plt.show()
```

# Histogram

```python
plt.figure(figsize=(10, 5))
plt.hist(df['Stock_Price'], bins=5, color='purple', alpha=0.7, edgecolor='black')
plt.xlabel('Stock Price')
plt.ylabel('Frequency')
plt.title('Stock Price Distribution')
plt.grid(True)
plt.show()
```

# Line plot

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Company_Name'], df['Stock_Price'], marker='o', linestyle='-', color='red')
plt.xlabel('Company Name')
plt.ylabel('Stock Price')
plt.title('Stock Price Trend')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
